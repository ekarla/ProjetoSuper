```python
from pylab import *
%matplotlib inline
```

```python
file_dados = 'dadosCovid.csv'
fd = pd.read_csv(file_dados)
fd.plot(x ='epidemiological_week', y ='last_available_confirmed',title = "CASOS CONFIRMADOS EM MANAUS")

```

```python
ax = plt.gca()

ax.set_title("Casos Confirmados X Mortes")
fd.plot(kind = 'line', x = 'epidemiological_week', y ='last_available_confirmed', ax=ax)
fd.plot(kind = 'line', x = 'epidemiological_week', y = 'last_available_deaths',color = 'red',ax=ax)
ax.set_xlabel('Semana Epidemiologica')
ax.set_ylabel('Quantidade')
plt.show()
```

<h4>GERANDO GRAFICO POR GRUPO</h4>

```python
# grafico mostra os municipios com casos confirmados
df.groupby('city')['new_confirmed'].nunique().plot(kind = 'bar', figsize = (20,15),title = 'Casos Confirmados por Municipio')
plt.show()
```

```python
colunas = ['epidemiological_week', 'city','last_available_confirmed','new_deaths']
```

```python
dadosSelecionados = df.filter(items = colunas)
```

```python
dadosSelecionados
```

```python
dadosSelecionados = dadosSelecionados.dropna()
```

```python
dadosSelecionados
```

```python
dadosSelecionados.groupby('city').count()
```

```python
#mortes confirmadas por municipio
df.groupby('city')['new_deaths'].nunique().plot(kind = 'bar', figsize = (20,15),title = 'Casos Confirmados por Municipio')
plt.show()
```
